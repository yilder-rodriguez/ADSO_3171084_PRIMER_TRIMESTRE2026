```python
class vehiculo: # clase padre
  def mover(self): #metodo con self
    print("el vehiculo se esta moviendo")

class carro(vehiculo): # clase hija (se agrego :)
  pass # ni atributos ni metodos (vacia)

#===========================================

def main(): # experiencia de usuario
  print("CLASE PADRE Y SU METODO\n")
  vehiculo1 = vehiculo() # Corregido a minúscula
  vehiculo1.mover() #invoca el metodo mover()

  print("Utilizaremos el metodo mover() de la clase padre en la clase HIJO\n")
  carro1 = carro()
  carro1.mover() # Se usa la instancia carro1

if __name__ == "__main__": # Se agrego :
  main()
```
