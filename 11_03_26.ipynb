```python
from abc import ABC, abstractmethod
# Crear la clase abstracta
class Figura(ABC):
    @abstractmethod
    def calcular_area(self):
        pass

# Crearemos una subclase
class circulo(Figura):
    def __init__(self,radio):
        self.radio = radio
    def calcular_area(self):
        return 3.1416 * (self.radio ** 2)
# **********************************************
class triangulo(Figura):
    def __init__(self,base,altura):
        self.base = base
        self.altura = altura
    def calcular_area(self):
        return (self.base * self.altura) / 2
# **********************************************
class rectangulo(Figura):
    def __init__(self,base,altura):
        self.base = base
        self.altura = altura
    def calcular_area(self):
        return self.base * self.altura
# **********************************************
class cubo(Figura):
    def __init__(self,lado):
        self.lado = lado
    def calcular_area(self):
        return 6 * (self.lado ** 3)
# **********************************************
class cilindro(Figura):
    def __init__(self,radio,altura):
        self.radio = radio
        self.altura = altura
    def calcular_area(self):
        return 3.1416 * (self.radio ** 2) * self.altura


 # def Calcular volumen
def mostrar_area(figura: Figura):
            print(f"Area: {figura.calcular_area():,.2f}")

def mostrar_base(figura: Figura):
            print(f"Base: {figura.calcular_area():,.2f}")
def mostrar_altura(figura: Figura):
            print(f"Altura: {figura.calcular_area():,.2f}")

def mostrar_volumen(figura: Figura):
            print(f"Volumen: {figura.calcular_area():,.2f}")
def mostrar_lado(figura: Figura):
            print(f"Lado: {figura.calcular_area():,.3f}")

def mostrar_radio(figura: Figura):
            print(f"Radio: {figura.calcular_area():,.2f}")

 # Ejecucion del programa
def main():
    circulo1 = circulo(5)
    mostrar_area(circulo(5)
    )
    triangulo1 = triangulo(5,7)
    mostrar_base(triangulo(5,7))
    mostrar_altura(triangulo(5,7))

    rectangulo1 = rectangulo(5,7)
    mostrar_area(rectangulo(5,7))
    mostrar_base(rectangulo(5,7))

    cubo1 = cubo(5)
    mostrar_lado(cubo(5))

    cilindro1 = cilindro(5,7)
    mostrar_radio(cilindro(5,7))
    mostrar_altura(cilindro(5,7))

if __name__ == "__main__":
     main()
```
