```python
class Registro:
        def __init__(self, nombre, apellido, documento, edad, programa, promedio):
            self.__nombre = nombre
            self.__apellido = apellido
            self.__documento = documento
            self.__edad = edad
            self.__programa = programa
            self.__promedio = promedio

  # Mostrar atributos

    # Nombre
        def get_nombre(self):
            return self.__nombre

    # apellido
        def get_apellido(self):
            return self.__apellido

    # Documento
        def get_documento(self):
            return self.__documento

    # Edad
        def get_edad(self):
            return self.__edad

    # Programa
        def get_programa(self):
            return self.__programa

    # Promedio
        def get_promedio(self):
            return self.__promedio

# metodos para modificar

      # Nombre

        def set_nombre(self, nombre):
          if isinstance(nombre, str):
            self.__nombre = nombre
          else:
            print("Debe ingresar un nombre valido")

      # Apellido

        def set_apellido(self, nuevo_apellido):
          if isinstance(nuevo_apellido, str):
            self.__apellido = nuevo_apellido
          else:
            print("Debe ingresar un apellido valido")

      # Documento

        def set_documento(self, nuevo_documento):
          if isinstance(nuevo_documento, int) and nuevo_documento >= 0:
             self.__documento = nuevo_documento
          else:
            print("Debe ingresar un documento valido")

      # Edad

        def set_edad(self, nueva_edad):
          if isinstance(nueva_edad, int) and nueva_edad >= 0:
            self.__edad = nueva_edad
          else:
            print("Debe ingresar una edad valida")

    # Programa

        def set_programa(self, nuevo_programa):
          if isinstance(nuevo_programa, str):
            self.__programa = nuevo_programa
          else:
            print("Debe ingresar un programa valido")

    # Promedio

        def set_promedio(self, nuevo_promedio):
          if isinstance(nuevo_promedio, float) and nuevo_promedio >= 0:
            self.__promedio = nuevo_promedio
          else:
            print("Debe ingresar un promedio valido")

    # Mostrar

        def mostrar_inf(self):
          print(f"Nombre: {self.__nombre}, Apellido: {self.__apellido}, Documento: {self.__documento}, Edad: {self.__edad}, Programa: {self.__programa}, Promedio: {self.__promedio}")

def main():
    Estudiante1 = Registro("Petro", "Maduro", 10784562252, 20, "Analisis de datos", 3.5)
    Estudiante2 = Registro("Lola", "Mento", 1023458776, 21, "Actor", 4.0)
    Estudiante3 = Registro("Juan", "Perez", 1054847561, 22, "Programacion", 3.8)

    # Informacion
    Estudiante1.mostrar_inf()
    Estudiante2.mostrar_inf()
    Estudiante3.mostrar_inf()

   # Actualizacion nombre

    print(f"nombre actual: {Estudiante1.get_nombre()}, {Estudiante2.get_nombre()},  {Estudiante3.get_nombre()}" )

    Estudiante1.set_nombre("Juan")
    Estudiante2.set_nombre("Lola")
    Estudiante3.set_nombre("Pedro")

    print("Actualizacion nombre estudiante 1")
    print("Actualizar nombre estudiante 2")
    print("Actualizar nombre estudiante 3")

  # Actualizacion apellido

    print(f"apellido actual: {Estudiante1.get_apellido()}, {Estudiante2.get_apellido()}, {Estudiante3.get_apellido()}")

    Estudiante1.set_apellido("Maduro")
    Estudiante2.set_apellido("Mento")
    Estudiante3.set_apellido("Perez")

    print("Actualizacion apellido estudiante 1")
    print("Actualizar apellido estudiante 2")
    print("Actualizar apellido estudiante 3")

  # Actualizacion documento

    print(f"documento actual: {Estudiante1.get_documento()}, {Estudiante2.get_documento()}, {Estudiante3.get_documento()}")

    Estudiante1.set_documento(10784562252)
    Estudiante2.set_documento(1023458776)
    Estudiante3.set_documento(1054847561)

    print("Actualizacion documento estudiante 1")
    print("Actualizar documento estudiante 2")
    print("Actualizar documento estudiante 3")

  # Actualizacion edad

    print(f"edad actual: {Estudiante1.get_edad()}, {Estudiante2.get_edad()}, {Estudiante3.get_edad()}")

    Estudiante1.set_edad(20)
    Estudiante2.set_edad(21)
    Estudiante3.set_edad(22)

    print("Actualizacion edad estudiante 1")
    print("Actualizar edad estudiante 2")
    print("Actualizar edad estudiante 3")

  # Actualizacion programa

    print(f"programa actual: {Estudiante1.get_programa()}, {Estudiante2.get_programa()}, {Estudiante3.get_programa()}")

    Estudiante1.set_programa("Programacion")
    Estudiante2.set_programa("Analisis de datos")
    Estudiante3.set_programa("Actor")

    print("Actualizacion programa estudiante 1")
    print("Actualizar programa estudiante 2")
    print("Actualizar programa estudiante 3")

  # Actualizacion promedio

    print(f"promedio actual: {Estudiante1.get_promedio()}, {Estudiante2.get_promedio()}, {Estudiante3.get_promedio()}")

    Estudiante1.set_promedio(3.5)
    Estudiante2.set_promedio(5.0)
    Estudiante3.set_promedio(4.5)

    print("Actualizacion promedio estudiante 1")
    print("Actualizar promedio estudiante 2")
    print("Actualizar promedio estudiante 3")

if __name__ == "__main__":
    main()
```
