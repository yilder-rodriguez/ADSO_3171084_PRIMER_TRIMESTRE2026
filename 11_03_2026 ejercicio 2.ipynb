```python
from abc import ABC, abstractmethod
# Clase abstracta con encapsulamiento y gestion de seguro
class Empleado(ABC):
  def __init__(self, nombre: str)-> None:
    # vamos a almacenar como privado para forzar el uso del getter y el setter
      self._nombre = None
      self.nombre = nombre # automaticamente llama el setter
    # GETTER
  @property
  def nombre(self) -> str:
    #acceso seguro al nombre del empleado
     return self._nombre
  @nombre.setter # setter profesional con validacion basica
  def nombre(self, valor: str) -> None:
    if isinstance(valor, str) and valor.strip():
      self._nombre = valor.strip()
    else:
      raise ValueError("El nombre debe ser un texto valido")
  @abstractmethod
  def trabajar(self) ->None:
    pass

# ahora crearemos a los hijos
class Gerente(Empleado):
  def trabajar(self) -> None:
      print(f"{self.nombre} Esta relacionado administracion de la empresa.")

class Desarrollador(Empleado):
  def trabajar(self) -> None:
    print(f"{self.nombre} Esta programando iso diagrama de clase en poo java.")
# polimorfismo en accion
def ejecutar_tarea(empleado: Empleado) -> None:
    empleado.trabajar()
# crearemos la ejecucion DICACTICA - MAIN
def main():
  print("************** INICIO DEL PROGRAMA EMPLEADOS ********************")
  gerente1 = Gerente("petro")
  desarrollador1 = Desarrollador("Uribe")
  ejecutar_tarea(gerente1)
  ejecutar_tarea(desarrollador1)
  print("/n**************************")
  print("Actualizando el nombre del desarrollo")
  desarrollador1.nombre = "Peppa pig"
  print(f"El nuevo desarrollador es {desarrollador1.nombre}")
  print("/n*************** Fin del programa  empleados")
if __name__ == "__main__":
  main()
```
